```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv("Iris.csv")
iris
```

```python
print("Shape:", iris.shape)

print("\nColumns:")
print(iris.columns)

print("\nFirst 5 Rows:")
print(iris.head())
```

```python
iris.info()

iris.describe()
```

```python
sns.scatterplot(
    data=iris,
    x='SepalLengthCm',
    y='PetalLengthCm',
    hue='Species'
)

plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python
iris.hist(figsize=(10,8))
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(data=iris)

plt.show()
```
